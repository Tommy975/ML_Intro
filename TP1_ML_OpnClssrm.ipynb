```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
# On charge le dataset dans un pandas dataframe
# 1. on store l'url de notre dataset dans une variable
# 2. on store notre dataset dans un pd df a l'aide de la méthode read_csv() de 
# la librairie pandas

url = "https://raw.githubusercontent.com/Tommy975/ML_DL/main/Datasets/house_data.csv?token=AFLKTMRULAZEVYKAG4NQILTADNBUG"
df = pd.read_csv(url)

# On utilise la méthode de la librairie pandas permettant de decrire notre jeu de données brutes
df.describe()
```

```
# Il y a quelques valeurs manquantes, on supprime ces lignes
df = df.dropna()

# Comme vu dans le TP, on a des outliers sur les grands propriétés
df1 = df[df["price"] < 10000]

# On reindexe
df1 = df1.reset_index(drop = True)
```

```
plt.plot(df1['surface'],df1['price'], 'x', markersize=4)
plt.show()
```

```
plt.plot(df1['arrondissement'], df1['price'], 'x', markersize=4)
```

```
X,Y = df1[['surface','arrondissement']], df1[['price']]


X,Y = X.values,Y.values


print("Shape de X : "+str(X.shape)) #Nous avons ici un array 2 dimensions.
print("Shape de Y : "+str(Y.shape)) 


```

```
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=.20)


```

```
regr = LinearRegression().fit(X_train,Y_train)
```

```
predicted = regr.predict(X_test)
```

```
M = regr.coef_
B = regr.intercept_
R_2 = regr.score(X_test,Y_test)
print("Les valeurs optimales M obtenues sont   : "+str(M))
print("La  valeur optimale B obtenue est       : "+str(B)[1:6])
print("La  valeur du R^2 obtenues est          : "+str(R_2)[0:4])
```
